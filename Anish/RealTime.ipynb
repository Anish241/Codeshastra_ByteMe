```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import threading
```

```python
model = load_model("../Bodhi/retrain_best_model.h5")
```

```python
class_labels = {0: "Incorrect", 1: "Correct"}
```

```python
def create_seq_of_frames(frames):
    
    # Determine the step size to select frames evenly
    processes_frames = []
    for frame_num in range(0,20):
        processes_frames.append(cv2.resize(frames[frame_num],(64,64))/255)
    return np.array(processes_frames)        
```

```python
def predict_on_frames(frames):
    processed_frames = create_seq_of_frames(frames)
    prediction = model.predict(np.array([processed_frames]))
    return class_labels[np.argmax(prediction)]
```
